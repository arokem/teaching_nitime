# Multi-channel analysis

In this example, we will look at neural data from cat area 17, recorded by Tim Blanche in the laboratory of Nicholas Swindale, University of British Columbia and downloaded from the NSF-funded [CRCNS data sharing web-site](http://crcns.org/data-sets/vc/pvc-3). 


This data has the advantage that the recordings were done using a polytrode:  




```python
from IPython.display import Image, display
display(Image(filename='./figures/polytrode.png'))
display(Image(filename='./figures/polytrode_in_brain.png'))
```

So, the data contains spikes from several different cells and we can look at bi-variate measures of coupling between the cells in the time and frequency domain

We start by importing some stuff we might need:

```python
import os
import nitime.timeseries as nts
import nitime.viz as viz
```

This is the folder that contains the data (I have downloaded it and put it in the same folder as the one I am running the notebook from):

```python
data_folder = './crcns_pvc3_cat_recordings/spont_activity/spike_data_area17/'
```

What's in there? 

```python
file_names1 = np.array(os.listdir(data_folder))
# The spike-time files are the ones with the '.spk' extension:
file_names2 = file_names1[np.array([f.endswith('.spk') for f in file_names1])]
```

```python
file_names2
```

Loop over the file-names (for time and memory reasons, we will only look at the first three channels here) and read in their content (`np.fromfile` is used to make an array from binary data, here in 64 bit integer format)

```python
ev = [] 
for f in file_names2:
    # Spike times are given in micro-second 
    ev.append(nts.Events(np.fromfile(os.path.join(data_folder, f), dtype=np.int64), time_unit='us'))        
```

Let's figure out how long the recording was:

```python
max_time = np.max(nts.TimeArray(nts.TimeArray(np.concatenate([e.time for e in ev]), time_unit=nts.base_unit), time_unit='s'))
print max_time
```

If we want to represent all the spikes in one big time-series, how large should it be? 

```python
t1 = nts.UniformTime(duration=max_time, sampling_interval=nts.TimeArray(1, time_unit='us'))
print t1.shape
```

Allocate an array with zeros. To save time and memory, we will also resample the data to msec resolution: 

```python
data = np.zeros((len(file_names2), t1.shape[0]/1000.))
```

We also need to know how to convert from the times (in micro-seconds) to indices into this array. This is simply the times divided by the sampling interval. 

```python
sampling_interval = t1.sampling_interval*1000.
print sampling_interval
print e.time[0]
print (e.time / sampling_interval)[0]
```

Nuke this memory hog:

```python
del t1
```

OK - let's add those 1's for the spikes. We convolve each spike train by 10 msec hamming window, so that we don't treat each spike as a 1 micro-second event

```python
for chan_idx, e in enumerate(ev):
    idx = e.time / sampling_interval
    times = idx[idx<data.shape[-1]]
    data[chan_idx, times.astype(int)] = 1
```

Let's look at the first 1% of the data (so that we don't run into time/memory issues). We set the sampling rate and keep the units as per default (seconds)

```python
ts1 = nts.TimeSeries(data, sampling_rate=10**3)
fig = viz.plot_tseries(ts1)
fig.set_size_inches([22, 4])
```

We import the analysis module to perform analysis on the data:

```python
import nitime.analysis as nta
```

Let's start by looking at a simple point-wise correlation in the time-domain: 

```python
correlation = nta.CorrelationAnalyzer(ts1)
```

The point-wise correlation between every pair of spike trains is very low: 

```python
fig = viz.drawmatrix_channels(correlation.corrcoef, [f.split('.')[0] for f in file_names2], color_anchor=[0, 1.0], cmap=matplotlib.cm.hot)
fig.set_size_inches([8,8])
```

Next, we can look at the coherency between the different channels. This is an analogue of correlation in the frequency domain: 
    
$R_{xy}(\lambda) = \frac{f_{xy}}{\sqrt{f_{xx}f_{yy}}}$
    

```python
C = nta.CoherenceAnalyzer(ts1, method=dict(NFFT=512))
```

$R_{xy}$ are complex quantities and we can derive two quantites from them. One is the coherence: 


$C_{xy}(\lambda) = \frac{|f_{xy}|^2}{f_{xx}f_{yy}}$


which is a measure of coupling strength in each frequency band: 

```python
alpha_idx = np.where(C.frequencies<10)
gamma_idx = np.where(np.logical_and(C.frequencies>25, C.frequencies<100))
fig = viz.drawmatrix_channels(np.mean(C.coherence[..., alpha_idx], -1).squeeze(), [f.split('.')[0] for f in file_names2], color_anchor=[0, 1.0], cmap=matplotlib.cm.hot)
fig.set_size_inches([8,8])
fig = viz.drawmatrix_channels(np.mean(C.coherence[..., gamma_idx], -1).squeeze(), [f.split('.')[0] for f in file_names2], color_anchor=[0, 1.0], cmap=matplotlib.cm.hot)
fig.set_size_inches([8,8])
```

Another is the angle of $R_{xy}$: 


$\phi_{xy} = ang(R_{xy})$


which is a measure of relative phase. 

```python
fig = viz.drawmatrix_channels(np.mean(C.phase[..., alpha_idx], -1).squeeze(), [f.split('.')[0] for f in file_names2], color_anchor=[-np.pi, pi])
fig.set_size_inches([8,8])
fig = viz.drawmatrix_channels(np.mean(C.phase[..., gamma_idx], -1).squeeze(), [f.split('.')[0] for f in file_names2], color_anchor=[-np.pi, pi])
fig.set_size_inches([8,8])
```
